```python
import pandas as pd
import webbrowser
import pyautogui
import time
import pyperclip

def abrir_whatsapp_web():
    """
    Abre o WhatsApp Web no navegador padrão.
    """
    webbrowser.open("https://web.whatsapp.com")
    time.sleep(5)  # Aguarda 15 segundos para o WhatsApp carregar completamente

def enviar_mensagens_reparo(df):
    """
    Envia mensagens personalizadas para uma lista de números no WhatsApp Web, com base nos dados de Reparo.
    
    :param df: DataFrame contendo os dados dos clientes
    """
    primeira_vez = True
    for index, row in df.iterrows():
        # Verificar se a célula contém "Sem Doc"
        if str(row['% Reparo']).strip().lower() == 'sem doc':
            print(f"Pulando linha {index} - % Reparo contém 'Sem Doc'")
            continue
        numero = row['Número']
        nome_loja = row['Nome Cliente']
        percent_reparo = row['% Reparo'] * 100  # Converte para porcentagem

        # Formatar a mensagem personalizada baseada no resultado do reparo
        if percent_reparo > 0:
            mensagem = (
                f"Boa tarde! Como você está?\n\n"
                f"Reparos de parabrisas de qualidade podem trazer grandes benefícios para sua loja, incluindo redução no tempo de serviço, menor chance de avaria e redução dos impactos do pós-atendimento, como descarte do vidro danificado.\n\n"
                f"RESULTADO DA {nome_loja}:\n"
                f"% Reparo de PB: {percent_reparo:.2f}%\n\n"
                f"Obrigado pelo seu apoio!\n\n"
                f"💡 Estratégias para Alavancar Seu Indicador:\n"
                f"1. Trinca Longa (Realização do reparo com a Resina Pit Filler) - https://youtu.be/jU2cHJsIleQ\n"
                f"2. Trinca Radial (Realização do reparo com a Resina Star Fill) - https://youtu.be/2izXl2-mW1Y\n"
                f"3. Expor ao segurado os benefícios do serviço, como isenção de franquia, redução no tempo de serviço e que irá manter a originalidade do veículo."
            )
        else:
            mensagem = (
                f"Boa tarde! Como você está?\n\n"
                f"Os reparos de qualidade podem trazer inúmeros benefícios para sua loja, como redução no tempo de serviço, menor chance de avaria e redução dos impactos do pós-atendimento, como descarte do vidro danificado.\n\n"
                f"RESULTADO DA {nome_loja}:\n"
                f"% Reparo: {percent_reparo:.2f}%\n\n"
                f"Você possui o equipamento necessário para realizar reparos? Oferecendo um serviço de reparo de qualidade, você pode melhorar a eficiência e satisfação dos seus clientes. Se precisar de mais informações ou tiver alguma dúvida, estou à disposição para ajudar!\n\n"
                f"Conto com o seu apoio!"
            )

        if primeira_vez:
            abrir_whatsapp_web()
            primeira_vez = False

        try:
            # Navega para o contato desejado
            pyautogui.hotkey('ctrl', 'alt', 'n')  # Abrir uma nova conversa
            time.sleep(0.5)
            pyautogui.typewrite(numero.replace('+', '').replace(' ', '').replace('-', ''))  # Digita o número do contato
            time.sleep(0.5)
            pyautogui.press('enter')

            # Copia e cola a mensagem
            pyperclip.copy(mensagem)
            time.sleep(0.5)  # Pausa necessária para garantir a cópia
            pyautogui.hotkey('ctrl', 'v')
            pyautogui.press("enter")
            print(f"Mensagem enviada para {numero}")

            # Pausa entre mensagens
            time.sleep(0.5)
        except Exception as e:
            print(f"Erro ao enviar mensagem para {numero}: {e}")

# Ler dados do Excel, começando da linha 9 e excluindo colunas extras
df = pd.read_excel(r'C:\Users\hugo.almeida\Desktop\ACOMPANHAMENTO HUGO (version 1).xlsm', sheet_name='ACOMPANHAMENTO', header=8)

# Verificar os primeiros registros do DataFrame
print(df.head())

# Enviar mensagens personalizadas
enviar_mensagens_reparo(df)
```
