```python
import pandas as pd
import webbrowser
import pyautogui
import time
import pyperclip

def abrir_whatsapp_web():
    """
    Abre o WhatsApp Web no navegador padr√£o.
    """
    webbrowser.open("https://web.whatsapp.com")
    time.sleep(5)  # Aguarda 15 segundos para o WhatsApp carregar completamente

def enviar_mensagens_reparo(df):
    """
    Envia mensagens personalizadas para uma lista de n√∫meros no WhatsApp Web, com base nos dados de Reparo.
    
    :param df: DataFrame contendo os dados dos clientes
    """
    primeira_vez = True
    for index, row in df.iterrows():
        # Verificar se a c√©lula cont√©m "Sem Doc"
        if str(row['% Reparo']).strip().lower() == 'sem doc':
            print(f"Pulando linha {index} - % Reparo cont√©m 'Sem Doc'")
            continue
        numero = row['N√∫mero']
        nome_loja = row['Nome Cliente']
        percent_reparo = row['% Reparo'] * 100  # Converte para porcentagem

        # Formatar a mensagem personalizada baseada no resultado do reparo
        if percent_reparo > 0:
            mensagem = (
                f"Boa tarde! Como voc√™ est√°?\n\n"
                f"Reparos de parabrisas de qualidade podem trazer grandes benef√≠cios para sua loja, incluindo redu√ß√£o no tempo de servi√ßo, menor chance de avaria e redu√ß√£o dos impactos do p√≥s-atendimento, como descarte do vidro danificado.\n\n"
                f"RESULTADO DA {nome_loja}:\n"
                f"% Reparo de PB: {percent_reparo:.2f}%\n\n"
                f"Obrigado pelo seu apoio!\n\n"
                f"üí° Estrat√©gias para Alavancar Seu Indicador:\n"
                f"1. Trinca Longa (Realiza√ß√£o do reparo com a Resina Pit Filler) - https://youtu.be/jU2cHJsIleQ\n"
                f"2. Trinca Radial (Realiza√ß√£o do reparo com a Resina Star Fill) - https://youtu.be/2izXl2-mW1Y\n"
                f"3. Expor ao segurado os benef√≠cios do servi√ßo, como isen√ß√£o de franquia, redu√ß√£o no tempo de servi√ßo e que ir√° manter a originalidade do ve√≠culo."
            )
        else:
            mensagem = (
                f"Boa tarde! Como voc√™ est√°?\n\n"
                f"Os reparos de qualidade podem trazer in√∫meros benef√≠cios para sua loja, como redu√ß√£o no tempo de servi√ßo, menor chance de avaria e redu√ß√£o dos impactos do p√≥s-atendimento, como descarte do vidro danificado.\n\n"
                f"RESULTADO DA {nome_loja}:\n"
                f"% Reparo: {percent_reparo:.2f}%\n\n"
                f"Voc√™ possui o equipamento necess√°rio para realizar reparos? Oferecendo um servi√ßo de reparo de qualidade, voc√™ pode melhorar a efici√™ncia e satisfa√ß√£o dos seus clientes. Se precisar de mais informa√ß√µes ou tiver alguma d√∫vida, estou √† disposi√ß√£o para ajudar!\n\n"
                f"Conto com o seu apoio!"
            )

        if primeira_vez:
            abrir_whatsapp_web()
            primeira_vez = False

        try:
            # Navega para o contato desejado
            pyautogui.hotkey('ctrl', 'alt', 'n')  # Abrir uma nova conversa
            time.sleep(0.5)
            pyautogui.typewrite(numero.replace('+', '').replace(' ', '').replace('-', ''))  # Digita o n√∫mero do contato
            time.sleep(0.5)
            pyautogui.press('enter')

            # Copia e cola a mensagem
            pyperclip.copy(mensagem)
            time.sleep(0.5)  # Pausa necess√°ria para garantir a c√≥pia
            pyautogui.hotkey('ctrl', 'v')
            pyautogui.press("enter")
            print(f"Mensagem enviada para {numero}")

            # Pausa entre mensagens
            time.sleep(0.5)
        except Exception as e:
            print(f"Erro ao enviar mensagem para {numero}: {e}")

# Ler dados do Excel, come√ßando da linha 9 e excluindo colunas extras
df = pd.read_excel(r'C:\Users\hugo.almeida\Desktop\ACOMPANHAMENTO HUGO (version 1).xlsm', sheet_name='ACOMPANHAMENTO', header=8)

# Verificar os primeiros registros do DataFrame
print(df.head())

# Enviar mensagens personalizadas
enviar_mensagens_reparo(df)
```
